## jieba

jieba分词的语料是基于人民日报。

安装命令：pip install jieba

版本：0.42.1

jieba的使用主要有几个方法：

    1. jieba.cut：生成器
    2. jieba.cut_for_search：生成器
    3. jieb.lcut：返回列表
    4. jieba.lcut_for_search：返回列表
   

```python
import jieba
str1 ='''时辈已争先，吾兄未著鞭。空嗟镊须日，犹是屈腰年。
不以殊方远，仍论水地偏。襄橙随客路，汉竹引归船。云木巴东峡，林泉岘北川。池馀骑马处，宅似卧龙边。
夜簟千峰月，朝窗万井烟。朱荷江女院，青稻楚人田。县舍多潇洒，城楼入醉眠。黄苞柑正熟，红缕鲙仍鲜。
坐厌牵丝倦，因从解绶旋。初辞五斗米，唯奉一囊钱。室好生虚白，书耽守太玄。枥中嘶款段，阶下引潺湲。
落照渊明柳，春风叔夜弦。绛纱儒客帐，丹诀羽人篇。雅论承安石，新诗与惠连。兴清湖见底，襟豁雾开天。
魏阙心犹系，周才道岂捐。一丘无自逸，三府会招贤。黄阁开帷幄，丹墀侍冕旒。位高汤左相，权总汉诸侯。
不改周南化，仍分赵北忧。双旌过易水，千骑入幽州。'''
word_object = jieba.cut(str1)  # 生成一个生成器对象
s=list(word_object)
print(s)
```

**jieba.cut 参数简介**
 
cut(self, sentence, cut_all=False, HMM=True) method of jieba.Tokenizer instance
    The main function that segments an entire sentence that contains
    Chinese characters into seperated words.
    
    Parameter:
        - sentence: The str(unicode) to be segmented.   （需要分词的字符串）
        - cut_all: Model type. True for full pattern, False for accurate pattern. （cut_all字段是分词的全模式和普通模式/默认模式）
        - HMM: Whether to use the Hidden Markov Model. （是否使用HMM模式分词，default True）

**名词解释**：

1.隐马尔可夫模型(Hidden Markov model, HMM)是一种结构最简单的动态贝叶斯网的生成模型，它也是一种著名的有向图模型。它是典型的自然语言中处理标注问题的统计机器学模型。
 
2.全模式和默认模式。

   str1= u“江州市长江大桥”为例： 
   
   jieba.cut(str1, cut_all=True)  #全模式是将词可能生成的词完全切分出来
   
　　Output：
        江州/ 州市/市长/长江/长江大桥/大桥  （可以理解为：所有可以切分成词的都会显示出来， 尽可能多得分词）
   
   jieba.cut(str1, cut_all=False)  

　　Output:
         江州/市/长江大桥



jieba.cut_for_search：该方法接受两个参数：
　　参数1：需要分词的字符串；
　　参数2：是否使用HMM模型，
该方法适用于搜索引擎构建倒排索引的分词，粒度比较细。

## Thulac

Thulac是清华大学的分词库，也是基于HMM的分词器。

安装命令：pip install thulac

版本：0.2.1

```python
import thulac
str1 ='''时辈已争先，吾兄未著鞭。空嗟镊须日，犹是屈腰年。
不以殊方远，仍论水地偏。襄橙随客路，汉竹引归船。云木巴东峡，林泉岘北川。池馀骑马处，宅似卧龙边。
夜簟千峰月，朝窗万井烟。朱荷江女院，青稻楚人田。县舍多潇洒，城楼入醉眠。黄苞柑正熟，红缕鲙仍鲜。
坐厌牵丝倦，因从解绶旋。初辞五斗米，唯奉一囊钱。室好生虚白，书耽守太玄。枥中嘶款段，阶下引潺湲。
落照渊明柳，春风叔夜弦。绛纱儒客帐，丹诀羽人篇。雅论承安石，新诗与惠连。兴清湖见底，襟豁雾开天。
魏阙心犹系，周才道岂捐。一丘无自逸，三府会招贤。黄阁开帷幄，丹墀侍冕旒。位高汤左相，权总汉诸侯。
不改周南化，仍分赵北忧。双旌过易水，千骑入幽州。'''
thu = thulac.thulac()
word = thu.cut(str1,  text=False)
print(word)
```

返回的分词结果是一个二维的列表（list），会显示分词结果和词性。

## snownlp

snownlp是基于3-gram的词性标注、HMM模型的分词算法。

安装命令：pip install snownlp
版本：0.12.3

```python
from snownlp import SnowNLP
str1 = '''时辈已争先，吾兄未著鞭。空嗟镊须日，犹是屈腰年。
不以殊方远，仍论水地偏。襄橙随客路，汉竹引归船。云木巴东峡，林泉岘北川。池馀骑马处，宅似卧龙边。
夜簟千峰月，朝窗万井烟。朱荷江女院，青稻楚人田。县舍多潇洒，城楼入醉眠。黄苞柑正熟，红缕鲙仍鲜。
坐厌牵丝倦，因从解绶旋。初辞五斗米，唯奉一囊钱。室好生虚白，书耽守太玄。枥中嘶款段，阶下引潺湲。
落照渊明柳，春风叔夜弦。绛纱儒客帐，丹诀羽人篇。雅论承安石，新诗与惠连。兴清湖见底，襟豁雾开天。
魏阙心犹系，周才道岂捐。一丘无自逸，三府会招贤。黄阁开帷幄，丹墀侍冕旒。位高汤左相，权总汉诸侯。
不改周南化，仍分赵北忧。双旌过易水，千骑入幽州。'''
snow_obj = SnowNLP(str1)
print(snow_obj.words)  # list 分词后返回一个分词结果的list
```

SnowNLP除了words属性外，还有如下几个：

    - snow_obj.tags #  一维list，内含元组（词，词性）
    - snow_obj.sentiments # positive的概率
    - snow_obj.pinyin # 每个词转为拼音，每个词构成list
